```python
def Return_sSMF(z_in, Fit = "G18"):
    if Fit == "G18":
        z_bin = np.digitize(z_in, z)
        SMF_Y = SMF_G18[z_bin]
        return SMF_MassRange_G18, SMF_Y
    elif Fit == "NoEvo":
        z_bin = np.digitize(z_in, z)
        SMF_Y = SMF_NoEvo[z_bin]
        return SMF_MassRange_NoEvo, SMF_Y
    else:
        #fit not used in model yet just use G18
        z_bin = np.digitize(z_in, z)
        SMF_Y = SMF_G18[z_bin]
        return SMF_MassRange_G18, SMF_Y

def Return_Cent_SMF(z_in, Fit = "G18"):
    if Fit == "G18":
        AbnMtch['Moster'] = False
        AbnMtch['G18'] = True
        AbnMtch['z_Evo'] = True
    elif Fit == "NoEvo":
        AbnMtch['Moster'] = False
        AbnMtch['G18'] = True
        AbnMtch['z_Evo'] = False
    elif Fit == "M13":
        AbnMtch['Moster'] = True
        AbnMtch['G18'] = False
        AbnMtch['z_Evo'] = True
    elif Fit == "B18":
        AbnMtch['Moster'] = False
        AbnMtch['G18'] = False
        AbnMtch['z_Evo'] = True
        AbnMtch['Behroozi18'] = True
        AbnMtch['B18c'] = True
    
    HM_Range = np.arange(9, 15, 0.1)#Virial   
    sSMF_X, sSMF_Y = Return_sSMF(z_in, Fit = Fit)
    if (Fit == "G18") or (Fit == "NoEvo"):
        cSMF_X, cSMF_Y = F.DM_to_SM(sSMF_X, np.log10(HMF_fun(HM_Range, z_in)), HM_Range, 0.1, 0.1, Paramaters, Redshift = z_in)
    elif (Fit == "M13") or (Fit == "B18"):
        DM_IN = np.log10(massdefs.changeMassDefinitionCModel(M=np.power(10, HM_Range), z=0.1, mdef_in='vir', mdef_out='200c')[0])
        if (Fit == "M13"):
            cSMF_X, cSMF_Y = F.DM_to_SM(sSMF_X, np.log10(HMF_fun(HM_Range, z_in)), DM_IN, 0.1, 0.1, Paramaters, Redshift = z_in)
        elif(Fit == "B18"):
            cSMF_X, cSMF_Y = F.DM_to_SM(sSMF_X, np.log10(HMF_fun(HM_Range, z_in)), DM_IN, 0.1, 0.1, Paramaters, Redshift = z_in, UseAlt = True)    
    #Reset Abn Parameters
    AbnMtch['z_Evo'] = True
    AbnMtch['Moster'] = False
    AbnMtch['G18'] = True
    AbnMtch['Behroozi18'] = False
    AbnMtch['B18c'] = False
    return sSMF_X, cSMF_Y

def Return_Total_SMF(z_in, Fit = "G18"):
    if Fit == "G18":
        AbnMtch['Moster'] = False
        AbnMtch['G18'] = True
        AbnMtch['z_Evo'] = True
    elif Fit == "NoEvo":
        AbnMtch['Moster'] = False
        AbnMtch['G18'] = True
        AbnMtch['z_Evo'] = False
    elif Fit == "M13":
        AbnMtch['Moster'] = True
        AbnMtch['G18'] = False
        AbnMtch['z_Evo'] = True

    sSMF_X, sSMF_Y = Return_sSMF(z_in, Fit = Fit)
    cSMF_X, cSMF_Y = Return_Cent_SMF(z_in, Fit = Fit)
    tSMF_Y = np.log10(sSMF_Y + np.power(10, cSMF_Y))
    #Reset Abn Parameters
    AbnMtch['z_Evo'] = True
    AbnMtch['Moster'] = False
    AbnMtch['G18'] = True
    return sSMF_X, tSMF_Y

def Return_Total_SMF_Alt(z_in, Fit = 'G18'):
    if Fit == "G18":
        AbnMtch['Moster'] = False
        AbnMtch['G18'] = True
        AbnMtch['z_Evo'] = True
    elif Fit == "NoEvo":
        AbnMtch['Moster'] = False
        AbnMtch['G18'] = True
        AbnMtch['z_Evo'] = False
    elif Fit == "M13":
        AbnMtch['Moster'] = True
        AbnMtch['G18'] = False
        AbnMtch['z_Evo'] = True
    
    HM_Range = np.arange(9, 15, 0.1)#Virial   
    sSMF_X, sSMF_Y = Return_sSMF(z_in, Fit = Fit)
    if Fit == "G18":
        cSMF_X, cSMF_Y = F.DM_to_SM(sSMF_X, np.log10(tHMF), SubMass, 0.1, 0.1, Paramaters, Redshift = z_in)
    elif Fit == "NoEvo": 
        cSMF_X, cSMF_Y = F.DM_to_SM(sSMF_X, np.log10(tHMF), np.log10(massdefs.changeMassDefinitionCModel(M=np.power(10, SubMass), z=0.1, mdef_in='vir', mdef_out='200c')[0]), 0.1, 0.1, Paramaters, Redshift = z_in)    
    #Reset Abn Parameters
    AbnMtch['z_Evo'] = True
    AbnMtch['Moster'] = False
    AbnMtch['G18'] = True
    return sSMF_X, cSMF_Y

f, axarr = plt.subplots(1, 2, figsize=(10,4))
for i in [0.1,0.5,1,1.5,2,2.5,3]:
    SMF_X, SMF_Y = Return_Total_SMF(i)
    axarr[0].plot(SMF_X, SMF_Y, label = i)
    SMF_X, SMF_Y = Return_sSMF(i)
    axarr[1].plot(SMF_X, np.log10(SMF_Y), "-.", label = i)
plt.legend()
plt.show()

f, axarr = plt.subplots(1, 2, figsize=(10,4))
for i in [0.1,0.5,1,1.5,2,2.5,3]:
    SMF_X, SMF_Y = Return_Total_SMF(i, Fit = "NoEvo")
    axarr[0].plot(SMF_X, SMF_Y, label = i)
    SMF_X, SMF_Y = Return_sSMF(i, Fit = "NoEvo")
    axarr[1].plot(SMF_X, np.log10(SMF_Y), "-.", label = i)
plt.legend()
plt.show()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
mpl.rcParams.update({'lines.linewidth': 1.5})
mpl.rcParams.update({'lines.markersize': 5})

def Shank17AM(HM, ScatterOn = False):
    Mstar0 = 10.68
    Mhalo0 = 11.8
    alpha = 2.13
    beta = 1.68
    
    Part1 = np.power(10, alpha*(HM-Mhalo0))
    Part2 = np.power(1+np.power(10,beta*(HM-Mhalo0)), -1)
    if ScatterOn:
        Scatter = np.random.normal(scale = 0.15, size = np.shape(HM))
        return Mstar0 + np.log10(Part1*Part2) + Scatter
    else:
        return Mstar0 + np.log10(Part1*Part2)
def DM_to_SM_for_S17(SMF_X, HMF, Halo_MR, HMF_Bin, SMF_Bin, Paramaters, Redshift = 0, N = 5000):
    """
    Input
    SMF_X: Stellar Mass Function Mass Range log10[$M_\odot$]
    HMF: Halo Mass Function Weights log10[ $\Phi$ Mpc^{-3} h^3]
    Halo_MR: Halo Mass Function Mass Range log10[$M_\odot$ h^{-1}]
    HMF_Bin: Binwidth of Halo_MR
    SMF_Bin: Binwidth of SMF_X
    N: number of times to use
    """

    DM_In = np.repeat(Halo_MR - np.log10(h), N) #log Mh [Msun]
    Wt = np.repeat(np.divide(np.power(10, HMF + 3*np.log10(h))*HMF_Bin, N), N) #Phi/N [Mpc^-3]
    
    SM = Shank17AM(DM_In, ScatterOn = True) #log M* [Msun]
    
    #SM = np.log10(HM_SM.mean_stellar_mass(prim_haloprop = 10**(DM_In + np.log10(0.704)) )) - np.log10(0.704) + np.random.normal(scale = 0.1, size = np.shape(DM_In))
    
    SMF_Y, Bin_Edge = np.histogram(SM, bins = np.append(SMF_X, SMF_X[-1]+SMF_Bin), weights = Wt) #Phi [Mpc^-3], M* [Msun]
    
    return SMF_X, np.log10(np.divide(SMF_Y, SMF_Bin)) #M* [Msun], Phi [Mpc^-3]

X = np.arange(10, 15, 0.1) #M h-1 #virial
f, axarr = plt.subplots(1, 2, figsize=(10,4))
Cent, Sat, Tot, leg, SDSS_Bin, Y_Sat_SDSS = Add_SDSS.SMF()
lines_plotted = []


Central = Override['M10'], Override['SHMnorm10'], Override['beta10'], Override['gamma10'], AbnMtch['Scatter']
Override['M10'], Override['SHMnorm10'], Override['beta10'], Override['gamma10'], AbnMtch['Scatter'] = 11.89,   0.031,   1.77,   0.52,   0.10
sSMF_X, discard = Return_sSMF(0.1, Fit = "G18")
axarr[0].plot(np.arange(10, 15, 0.1)- np.log10(h), F.DarkMatterToStellarMass(np.arange(10, 15, 0.1) - np.log10(h), 0.1, Paramaters, ScatterOn = False), "ok",label = "G18", mfc='none')
totSMF_X, totSMF_Y = F.DM_to_SM(sSMF_X, np.log10(tHMF), SubMass, 0.1, 0.1, Paramaters, Redshift = 0.1)
lines_plotted.append(axarr[1].plot(totSMF_X, totSMF_Y, "-k", label = "G18 Total")[0])
Override['M10'], Override['SHMnorm10'], Override['beta10'], Override['gamma10'], Override['Scatter'] = Central


AbnMtch['Behroozi18'] = True
AbnMtch['B18t'] = True
axarr[0].plot(X- np.log10(h), F.DarkMatterToStellarMass( X - np.log10(h), 0.1, Paramaters, ScatterOn = False), "pC4",label = "B18", mfc = 'none')
AbnMtch['B18t'] = False
AbnMtch['Behroozi18'] = False

axarr[0].plot(X- np.log10(h), Shank17AM(np.log10(massdefs.changeMassDefinitionCModel(M=np.power(10, X), z=0.1, mdef_in='vir', mdef_out='200c')[0]) - np.log10(h)), "vC8",label = "S17", mfc = 'none')
AbnMtch['Moster'] = True
AbnMtch['G18'] = False
axarr[0].plot(X- np.log10(h), F.DarkMatterToStellarMass(np.log10(massdefs.changeMassDefinitionCModel(M=np.power(10, X), z=0.1, mdef_in='vir', mdef_out='200c')[0]) - np.log10(h), 0.1, Paramaters, ScatterOn = False), "dC1",label = "M13", mfc = 'none')
AbnMtch['Moster'] = False
AbnMtch['G18'] = True
axarr[0].plot(X- np.log10(h), F.DarkMatterToStellarMass(X - np.log10(h), 0.1, Paramaters, ScatterOn = False), "-b", label = "G18")


X, Y = Return_Cent_SMF(0.1, Fit = "B18")
lines_plotted.append(axarr[1].plot(X, Y, "-.C4",label = "B18 Central")[0])
X, Y = Return_Cent_SMF(0.1, Fit = "M13")
lines_plotted.append(axarr[1].plot(X, Y, "-.C1",label = "M13 Central")[0])
X, Y = Return_Cent_SMF(0.1, Fit = "G18")
lines_plotted.append(axarr[1].plot(X, Y, "-b",label = "G18 Central")[0])
sSMF_X, discard = Return_sSMF(0.1, Fit = "G18")
totSMF_X, totSMF_Y = DM_to_SM_for_S17(sSMF_X, np.log10(cHMF), np.log10(massdefs.changeMassDefinitionCModel(M=np.power(10, SubMass), z=0.1, mdef_in='vir', mdef_out='200c')[0]), 0.1, 0.1, Paramaters, Redshift = 0.1)
lines_plotted.append(axarr[1].plot(totSMF_X, totSMF_Y, ":C8", label = "S17 Central")[0])


AbnMtch['Behroozi18'] = True
AbnMtch['B18c'] = True
axarr[0].plot(np.arange(10, 15, 0.1)- np.log10(h), F.DarkMatterToStellarMass_Alt(np.arange(10, 15, 0.1) - np.log10(h), 0.1, Paramaters, ScatterOn = False), "-.C4",label = "B18")
AbnMtch['B18c'] = False
AbnMtch['Behroozi18'] = False
mpl.rcParams.update({'font.size': 13})
axarr[0].legend(loc = 4, frameon = False)
axarr[1].legend(handles=lines_plotted, loc = 3, frameon = False)

axarr[0].set_ylim(9, 12.5)
axarr[0].set_xlim(11, 15)
axarr[0].set_xlabel("$\mathrm{log_{10}}$ $\mathrm{M_h}$ $\mathrm{[M_\odot]}$", fontproperties = mpl.font_manager.FontProperties(size = 13))
axarr[0].set_ylabel("$\mathrm{log_{10}}$ $\mathrm{M_*}$ $\mathrm{[M_\odot]}$", fontproperties = mpl.font_manager.FontProperties(size = 13))
axarr[1].set_xlabel("$\mathrm{log_{10}}$ $\mathrm{M_*}$ $\mathrm{[M_\odot]}$", fontproperties = mpl.font_manager.FontProperties(size = 13))
axarr[1].set_ylabel("$\mathrm{log_{10}}$ $\mathrm{\phi}$ $\mathrm{[Mpc^{-3}dex^{-1}]}$", fontproperties = mpl.font_manager.FontProperties(size = 13))
axarr[1].set_ylim(-7, -1)
plt.tight_layout()
plt.savefig("Figures_For_Paper_Final/Fig2.png")
plt.savefig("Figures_For_Paper_Final/Fig2.pdf")
plt.savefig("Figures_For_Paper_Final/Fig2.svg")

plt.show()
mpl.rcParams.update(mpl.rcParamsDefault)
```
