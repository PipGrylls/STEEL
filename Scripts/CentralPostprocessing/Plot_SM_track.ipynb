```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import StrMethodFormatter
plt.ion()
```

```python
file_name = 'SM_track_data/Berhoozi_SM_forFS.txt'
z_Beh = np.loadtxt(file_name, skiprows=3, usecols=0)
SM_Beh = np.loadtxt(file_name, skiprows=3, usecols=(1,2,3,4))
z = np.loadtxt('SM_track_data/Redshift_array.txt')
AvaSM_G19 = np.loadtxt('SM_track_data/SM_track_G19.txt')
AvaSM_B19 = np.loadtxt('SM_track_data/SM_track_B19.txt')
```

```python
plt.figure(figsize=(10,7))
k=0
for k in range(0,3):
    if k==0:
        plt.semilogx(z_Beh, SM_Beh[:,k], '-', color='red', label='Behroozi data')
        plt.semilogx(z, AvaSM_G19[:,k], '-.', color='purple', label='Grylls-19')
        plt.semilogx(z, AvaSM_B19[:,k]-0.05, '-.', color='black', label='Behroozi-19')
    else:
        plt.semilogx(z_Beh, SM_Beh[:,k], '-', color='red')
        if k==1:
            plt.semilogx(z, AvaSM_G19[:,k]-0.06, '-.', color='purple')
            plt.semilogx(z, AvaSM_B19[:,k]+0.01, '-.', color='black')
        else:
            plt.semilogx(z, AvaSM_G19[:,k], '-.', color='purple')
            plt.semilogx(z, AvaSM_B19[:,k], '-.', color='black')
plt.xlim(0.1,4)
plt.ylim(9.5, 11.5)
plt.xlabel(r'$z$', fontsize=20)
plt.ylabel(r'$log10 M_*$', fontsize=20)
plt.legend(fontsize=16)
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}'))
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}'))
mpl.rc('xtick', labelsize=15) 
mpl.rc('ytick', labelsize=15)

#plt.savefig('SM_track.pdf', dpi=300)
```
